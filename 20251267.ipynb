3-3. 정답 : 1, 3, 6, 7, 9, 8, 5  이유 : 3의 왼쪽 자식 끝 -> 1의 부모 -> 8의 왼쪽 서브트리 중 가장 아래 자식 -> 6의 부모 -> 8의 오른쪽 자식 -> 7과 9의 부모 -> 루트

3-4.   
    베이스 케이스: h = 1  
        - T는 단 한 개의 노드(루트)를 갖는다  
        -  유일한 노드를 방문하는 것은 당연히 정렬된 순서다  
    귀납적 가정: k < h인 모든 k에 대해 성립한다 가정하자  
    귀납적 전개: (정리가 k = h에 대해 성립함을 보인다)  
        -  이진 검색 트리 T는 오른쪽과 같이 그릴 수 있다  
        -  귀납적 가정에 의해 중위 순회는 TL과 TR을 각각 정렬된 순서로 방문한다  
        -  TL의 모든 키는 r의 키보다 작고, TR의 모든 키는 r의 키보다 크므로, 중위 순회 TL→ r → TR는 정렬된 순서다
